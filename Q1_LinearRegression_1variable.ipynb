```python
# import the required libraries
from LinearRegression import LR_1var
import random
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sn
```

### Load the data

```python
# read the data from csv using pandas
data = pd.read_csv("cancer_dataset_wpbc.csv")
```

### Analyze the data

```python
data[['Mean Texture','Tumor Size']].describe()
```

We can see the two columns count are same which doesn't there wot be nulls there could be a possibility that all the two columns might have same count of null values. so we perform the below step as a formal procedure to remove null rows if exists any.

### Drop null rows

```python
data = data.dropna(axis=0)
```

### Data Visualization using BoxPlot for data point analysis

```python
X = data['Mean Texture'].values
Y = data['Tumor Size'].values
plt.figure(figsize=(20, 4)) 

plt.subplot(1, 2, 1)
plt.boxplot(X, vert=False)
plt.title('Mean Texture')

plt.subplot(1, 2, 2)
plt.boxplot(Y, vert=False)  
plt.title('Tumor Size')

plt.show()
```

We can clearly see that there are outliers in both of the column data points. This suggests that the data may not be stable, which can potentially lead to worse performance in the metrics.

Also we can observe that the median line of "Mean Texture" is close to the left end of the box which means the data is left skewed.

### Data Visualization using Scatter Plot for data point analysis

```python
plt.scatter(X, Y)
plt.xlabel('Mean Texture')
plt.ylabel('Tumor Size')
plt.title('Mean Texture Data Vs Tumor Size')
plt.show()
```

We can clearly see that the data points are widespread and also the presence of outliers which indicates that the relationship between the two variables is complex and may not be well-described by a simple linear model.

The scatter of data points across a broad range of values indicates that there may not be a strong or clear linear relationship between the two variables

### Load the data into input and target variable respectively into an array of shape nX1

```python
X1 = np.array(X).reshape(-1,1)
Y = np.array(Y).reshape(-1,1)
```

### Train, Test and Prediction  Metrics

```python
model = LR_1var(data,X1,Y,0.7,0.004,100000,0.000001)
model.train()
model.predict()
```

### Observation

From the above metrics we can observe that R squared error is negative which means our model is a poor fit.

```python
model.plot_cost_history()
model.plot_regression()
```

### Cost function observation

When the cost function is going down during training, it means the model is getting better. It's like the model is trying to make fewer mistakes or reach its goal. But, just because the cost is decreasing, it doesn't mean the model is perfect. It could still have some room to improve. So, it's important to keep an eye on the cost to make sure the model is learning well and getting closer to what we want it to do.
