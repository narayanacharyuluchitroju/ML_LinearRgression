```python
# import the required libraries
from LinearRegression import LR_2var
import random
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sn
```

### Load the data

```python
# read the data from csv using pandas
data = pd.read_csv("cancer_dataset_wpbc.csv")
```


### Analyze the data

```python
data[['Mean Texture','Lymph Node Status','Tumor Size']].describe()
```

We can see the two columns count are same which doesn't there wot be nulls there could be a possibility that all the three columns might have same count of null values. so we perform the below step as a formal procedure to remove null rows if exists any.

### Drop null rows

```python
data = data.dropna(axis=0)
```

### Data Visualization using BoxPlot for data point analysis


```python
X1 = data['Mean Texture'].values
X2 = data['Lymph Node Status'].values
Y = data['Tumor Size'].values
plt.figure(figsize=(20, 6)) 

plt.subplot(3,1, 1)
plt.boxplot(X1, vert=False)
plt.title('Mean Texture')

plt.subplot(3, 1, 2)
plt.boxplot(X2, vert=False)  
plt.title('Lymph Node Status')

plt.subplot(3, 1, 3)
plt.boxplot(Y, vert=False)  
plt.title('Tumor Size')

plt.show()
```

For "Lymph Node Status" figure we can see the data is mostly left skewed and there are many outliers in the data points which will show great impact on data analysis.

The left whisker is very short and close to the edge of the plot, making it less visible. This means there may be no data points below the left whisker, which means the minimum value is very close to the first quartile (Q1) and there are no outliers on the lower end

### Data Visualization using Scatter Plot for data point analysis

```python
# Sample data (replace these with your data)
X1 =  np.array(data['Mean Texture'].values).reshape(-1,1)
X2 =  np.array(data['Lymph Node Status'].values).reshape(-1,1)
Y = np.array(data['Tumor Size'].values).reshape(-1,1)

# Create a scatter plot
plt.figure(figsize=(6, 6))  # Adjust figure size if needed
plt.scatter(X1, Y, label='Mean Texture vs Target Size', marker='o', c='blue', alpha=0.7)
plt.scatter(X2, Y, label='Lymph Node Status vs Target Size', marker='x', c='red', alpha=0.7)

# Customize plot
plt.title('Scatter Plot')
plt.xlabel('[Mean Texture, Lymph Node Status]')
plt.ylabel('Tumor Size')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

We can clearly see that the data points are widespread and also the presence of outliers which indicates that the relationship between the two variables is complex and may not be well-described by a simple linear model.

The scatter of data points across a broad range of values indicates that there may not be a strong or clear linear relationship between the three variables
